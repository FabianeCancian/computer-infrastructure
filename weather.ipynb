### Data Analysis
The analysis contains hourly weather observations from Athenry, Co Galway find on the weather.json file. The analysis focuses on exploring key weather variables like temperature, wind speed, and humidity. Trends and patterns in local weather conditions are identified through the data.

***

#### Dataset description:
"This file contains a list of observations for every hour of the current day for our synoptic station in Athenry, Co Galway. The file is updated hourly. Time values are Local times. Values for each hour include: Name; Dry bulb temperature in whole degrees; Weather description, Windspeed (kt); Cardinal Wind Direction; Relative Humidity (%); Rainfall (mm); msl Pressure (mbar); Day of the week; Date; Time of observation."

Source: https://data.gov.ie/dataset/todays-weather-athenry

````python
# Data frames 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load the dataset.
df = pd.read_json('data/weather/weather.json')
# Let's have a look
df
````

````python
# Describe the data set.
df.describe()
````

````python
#Counting how many hours each weather element last
df["weatherDescription"].value_counts()
````

````python
#Plot
weatherDescription = df['weatherDescription'].to_numpy()
plt.hist(weatherDescription, color='blue', edgecolor= 'black')
plt.title('Weather through time')
plt.xlabel("Weather")
plt.ylabel("Number of hours")
````

````python
#Plot the wind speed variance during the day.
plt.figure(figsize=(13.5, 4))
plt.xlabel("Hour of the day")
plt.ylabel("Wind Speed")
sns.lineplot(data=df, x='reportTime', y='windSpeed', color='#5a0a29').set(title='Windspeed Variance')
````

Wind speeds were mostly moderate, with a slight peak at 13 km/h, suggesting breezy but not extremely windy conditions.

````python
#Plot the Temperature variance during the day.
plt.figure(figsize=(13.5, 4))
plt.xlabel("Hour of the day")
plt.ylabel("Air humidity %")
sns.lineplot(data=df, x='reportTime', y='humidity', color='#6ca254').set(title='Air Humidity Percentage Along the Day')
````

The high average humidity (around 91%) suggests that the area was likely experiencing moist, possibly overcast or rainy conditions, though the actual rainfall was minim.

````python
#Plot the Temperature variance during the day.
plt.figure(figsize=(13.5, 4))
plt.xlabel("Hour of the day")
plt.ylabel("Temperature in C")
sns.lineplot(data=df, x='reportTime', y='temperature', color='#191cb6').set(title='Temperature Variance')
````

The average temperature is 9C, indicating cool conditions. This temperature range suggests mild, cool weather typical of early winter or late autumn. The temperature fluctuates between 6C and 11C, with only slight variations, implying a stable and relatively consistent climate.

In summary on November 17, 2024, the weather can be characterized by cool, stable conditions, with an average temperature of 9C, indicating mild fall or early winter weather. The temperature ranges from 6C to 11C, suggesting minimal fluctuation. Wind conditions are mild, with an average speed of 8 km/h coming from the west-southwest. While there is very light rainfall (0.08 mm),humidity levels are high at 91%, resulting in a damp atmosphere that can make the air feel cooler than the actual temperature, and could contribute to discomfort for those not accustomed to such conditions. Atmospheric pressure remains stable at 1016.6 hPa, suggesting that no major weather changes or storm systems are expected. Overall, the day will feature cool, moist conditions with light winds, high humidity, and minimal rain, providing a calm and settled environment, though the dampness in the air and light breeze may make it feel cooler than the actual temperature.
